- Below is some companies revenues in US dollars. Using numpy can you convert this into Pakistani rupees? 1 USD = 277 PKR

```python
import numpy as np
```

`Revenue in PKR`

```python
revenue_in_usd = np.array([[200, 220, 250], [68, 79, 105], [110, 140, 180], [80, 85, 90]])
exchange_rate = 277
revenue_in_pkr = revenue_in_usd * exchange_rate
revenue_in_pkr
```

- Divine flowers is a flower shop that sells different type of flowers. Below is the table showing how many flowers of each type they sold in different months. Also given are the prices of one flower each. Using this find out their total sales in every month.

`Total Sales`

```python
units_sold = np.array([[50, 60, 25], [10, 13, 5], [40, 70, 52]])
price_per_unit = np.array([20, 30,15])

total = np.dot(price_per_unit, units_sold)
total
```
